```python
'''
->Pass by value and pass by reference
    Python is neither purely pass-by-value nor pass-by-reference.
    Python uses “pass by object reference” (also called call by sharing).
'''
```

1.Pass by Value

```python
#passing values
def update(x):
    x = 10
    print("Inside function:", x)

a = 5
update(a)
print("Outside function:", a)
```

2.Pass by Reference

```python
#passing objects
def update(lst):
    lst.append(4)
    print("Inside function:", lst)

nums = [1, 2, 3]
update(nums)
print("Outside function:", nums)
```
