```python
#counting the number of times a number appears in an array

#brute force approach

def count_elements(arr):
    visited = [False]*len(arr)

    for i in range(len(arr)):
        if visited[i] is True:
            continue
        count = 1

        for j in range(i+1, len(arr)):
            if arr[i] == arr[j]:
                visited[j] = True
                count += 1

        print(arr[i],count)

arr = [1,2,3,4,1,2,3,4,5]
count_elements(arr)
```

```python
#Optimal Approach --> using Hash Maps

def hashing(arr):
    hash = {}

    for i in arr:     #iterate through each element in array
        if i in hash:   #if element already exist in hash ( dictionary ) -> increment it
            hash[i] += 1
        else:
            hash[i] = 1

    print(hash)


arr = [1,2,3,3,4,5,1,2,3]
hashing(arr)
```

```python
# Find the Frequency of the highest and the lowest frequency element using the concept of hashing

array = [10,5,10,15,10,5]
dict = {}
maxfreq = 0
maxkey = 0
minfreq = len(array)
minkey = 0

for i in array:
    if i in dict:
        dict[i] += 1
    else:
        dict[i] = 1

for key, values in dict.items():
    if maxfreq < values:
        maxfreq = values
        maxkey = key
    if minfreq > values:
        minfreq = values
        minkey = key

print(maxkey, minkey)
```
