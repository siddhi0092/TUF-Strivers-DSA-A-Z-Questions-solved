```python
#Recursion

# print name n times 
#complexity ->   O(n) for both
def recursion(n, name):
    if n == 0:
        return 
    print(name)
    recursion(n-1, name)
n = int(input("Enter no."))
name = input("Enter your name")
recursion(n,name)
```

```python
def recursion(n):
    if n == 0:
        return
    print(n)
    recursion(n-1)

n = 5
recursion(n)
```

```python
def recursion(current,n):
    if current>n:
        return 
    print(current)
    recursion(current + 1, n)
recursion(1,5)
    
```

```python
#Sum of first n numbers using recursion 

def recursion(n, sum):
    if n<1:
        print(sum)
        return
    recursion(n-1,sum +n)

recursion(6,0)
```

```python
#factorial

def recursion(n, fact):
    if n<1:
        print(fact)
        return
    recursion(n-1,fact*n)

recursion(5,1)
```

```python
#reversing an array using for loop

n = [1,2,3,4,5]
rev = []
for i in range(len(n)-1, -1, -1): 
    rev.append(n[i])

print(rev)
```

```python
#reversing an array using recursion

def rev(n, length, re):
    if length == 0:
        return re
    re.append(n[length-1])
    return rev(n, length-1,re)

n = [1,2,3,4,5]
re = []
length = len(n)
rev(n,length,re)
```

```python
#more effecient way (we do not need to create a new list every time)
#swapping

def rev(n, i = 0):
    if i >= len(n)//2:
        return n
    n[i], n[len(n)-1-i] = n[len(n)-1-i], n[i]
    return rev(n,i+1)

n = [1,2,3,4,5]
print(rev(n))
```

```python
#normal without using recursion 
def rev(n):
    if n == n[::-1]:
        return True
    else:
        return False

n = "ABCDCA"
if rev(n) == True:
    print("Is a palindrome")
else:
    print("is not a palindrome")
```

```python
#Palindrome using recursion

def ispalindrome(s, i = 0):
    if i >= len(s)//2:
        return True
    
    if s[i] != s[len(s)-i-1]:
        return False
    
    return ispalindrome(s,i+1)

s = "MADAM"
if ispalindrome(s) == True:
    print("Is a palindrome")
else:
    print("is not a palindrome")
```

```python
#Fibonacci Series 

def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1)+ fibonacci(n-2)

n= 6
print(fibonacci(n))
```
