```python
#Selection Sort

arr = [13,46,24,52,20,9]
for i in range(0, len(arr)):
    minid = i
    for j in range(i+1,len(arr)):
        if arr[minid] > arr[j]:
            minid = j
    if arr[minid] < arr[i]:
        arr[minid],arr[i] = arr[i], arr[minid]

print(arr)
```

```python
#Bubble sort

arr = [5,4,3,2,1]

for i in range(len(arr)):
    for j in range(i+1,len(arr)):
        if arr[i] > arr[j]:
            arr[i], arr[j] = arr[j], arr[i]  #swap

print(arr)
```
