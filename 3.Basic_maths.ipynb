```python
#1 .Counting the number of digits in a number 

#1.1 Brute force method 

a = int(input("Enter a number : "))
count = 0
while(a > 0):
    count = count + 1
    a = a // 10

print(count)

```

```python
#1.2 Optimal Solution  
# digits = ⌊log10​(n)⌋ + 1

# tells us what power of 10 gives n.

# Example:

# log₁₀(1) = 0

# log₁₀(10) = 1

# log₁₀(100) = 2

# log₁₀(1000) = 3

import math

n = int(input("Enter a number : "))

count = int(math.log10(n)+1)

print(count)
```

```python
# 3. Check Whether a number is a palindrome or not

n = int(input("Enter a number : "))

#first we will reverse a number and then compare and check if it a palindrome or not
original = n
rev = 0
dig = 0
while n>0:
    dig = n % 10
    rev = rev * 10 + dig
    n = n // 10

if rev == original:
    print("Number is a palindrome")
else:
    print("number is not a palindrome")
```

```python
# 4. Find the GCD of 2 numbers 
# 4.1 Brute force approach
import time

start_time = time.time()
n1 = int(input("Enter n1"))
n2 = int(input("Enter n2"))

mini = min(n1,n2)
ans = 1
for i in range(1,mini+1):
    if n1%i == 0 and n2%i == 0:
        ans = i
end_time = time.time()
total_time = end_time - start_time
print(ans)
print(total_time)
```

```python
#4.2 Better Approach
import time

start_time = time.time()
n1 = int(input("Enter n1"))
n2 = int(input("Enter n2"))

mini = min(n1,n2)
ans = 1
for i in range(mini+1, 1, -1):
    if n1%i == 0 and n2%i == 0:
        ans = i
end_time = time.time()
total_time = end_time - start_time
print(ans)
print(total_time)
```

```python
#4.3 Optimal Approach   --> Euclidean Algorithm
#Using Recursion

import time

start = time.time()
def gcd(n1,n2):
    maxi = max(n1,n2)
    mini = min(n1,n2)
    if maxi-mini == 0:
        return mini
    return gcd(maxi-mini, mini)

if __name__ == '__main__':
    n1 = int(input("Enter n1"))
    n2 = int(input("Enter n2"))
    print(gcd(n1,n2))
    end = time.time()
    total = end - start
    print(total)
```

```python
#6 Print all the divisors

#6.1 brute force approach

n = int(input("Enter a number"))
divisors = []
for i in range(1,n+1):
    if n % i == 0:
        divisors.append(i)

print(divisors)
```

```python
#6.2 Optimal Appraoch

"""We can optimise the previous approach by using the property that for any non-negative integer n, if d is a divisor of n then n/d is also a divisor of n. This property is symmetric about the square root of N. Thus, by traversing just the first half we can avoid redundant iteration and computations improving the efficiency of the algorithm."""

import math
import time 

start = time.time()
n = int(input("Enter a number"))
divisors = []

for i in range(1, int(math.sqrt(n))+1):
    if n % i == 0:
        divisors.append(i)
        if i != n // i:
            divisors.append(n//i)
divisors.sort()
print(divisors)
end = time.time()

total = end-start
print(total)
```

```python
#7. check if a number is prime or not

#7.1 Brute force approach

import time

start = time.time()
n = int(input("Enter a no."))

if n <= 1:
    print("not a prime no.")
elif n == 2:
    print("is a prime no.")
else:
    for i in range(2,n):
        if n % i == 0:
            print("not a prime no.")
            break
        else:
            print("it is a prime no.")
            break
end = time.time()
total = end-start

print(total)
```
