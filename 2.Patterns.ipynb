```python
#Important concept 
n = 5
for i in range(0, n):
    for a in range(0, n-i-1):   #space
        print("-", end="")
    for b in range(0, 2*i+1):   #star
        print("*",end="")
    for c in range(0, n-i-1):   #space
        print("-", end="")
    print("")
        
    
```
