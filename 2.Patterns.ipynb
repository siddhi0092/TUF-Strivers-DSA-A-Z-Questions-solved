```python
#binary triangle

n = 5
start = 1
for i in range(0,n):
    if i%2==0:
        start = 1
    else :
        start = 0
    for j in range(0,n):
        if j<=i:
            print(start, end="")
            start = 1 - start
    print("")
```

```python
n = 5
for i in range(1,n):
    for j in range(1,n):  #number pyramid 
        if j<=i:
            print(j, end="")
        else:
            print("-", end="")
    for j in range(1,n-1):   #space 
        if j>=i:
            print("-", end="")
    for j in range(n,0,-1):  #number pyramid
        if j<=i:
            print(j, end="")
    print("")
```

```python
#Alpha Pattern Hill
n = 5
for i in range(0,n-1):
    
    for j in range(n,0,-1):   #spaces
        if j>i+1:
            print("-", end="")
    for k in range(0,n):      #first pattern
        alpha = chr(ord('A')+k)
        if k<=i:
            print(alpha, end="")
    for l in range(n,-1,-1):      #2nd pattern
        alpha = chr(ord('A')+l)
        if l<i:
            print(alpha, end="")
    print()
```
